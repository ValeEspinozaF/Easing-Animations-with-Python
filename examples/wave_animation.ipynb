```python
%load_ext autoreload
%autoreload 2
from easing import easing
import numpy as np
import pandas as pd
```

```python
x_array = np.linspace(0, 2*np.pi, 100)

kfs = []

for i in np.linspace(0, 1, 5):
    y_array = np.sin(x_array/0.1 + 2*np.pi*i)*0.3

    keyframe_df = pd.DataFrame({
        "x": x_array,
        "y": y_array
    })

    kf_flattened = keyframe_df[["x", "y"]].to_numpy().reshape(1,-1)
    kfs.append(kf_flattened)

data_df = pd.DataFrame(np.vstack(kfs))
eased_data = easing.Eased(data_df, wrap=False)

eased_data.power_ease(n=1)

eased_data.polyline_animation2d(duration=2.0, 
                                feat_kws={'color':"#d62728"},
                                ax_kws={'xlim':(0, 2*np.pi), 'xlabel':'x', 'ylabel':'y', 'title':'polyline_animation2d()'},
                                #destination='media/wave.gif', 
                                figsize=(4,4),
                                )
```
